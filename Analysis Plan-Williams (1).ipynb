```python
plt.figure(figsize=(12, 6))   
counts = owls_df["region_name"].value_counts()
top20 = counts.head(20)
top20.plot(kind="bar", rot=45)
plt.title("Top 20 Regions")
plt.xlabel("Region Name")
plt.ylabel("Abundance Mean")
plt.tight_layout()
plt.show()
```

When sorted by region the data shows the highest regions located in North America, however there are several South American regions that are very prevelant in this data

### Forest Loss/Coverage Data


```python
url = "https://ourworldindata.org/grapher/annual-change-forest-area.csv?v=1&csvType=full&useColumnShortNames=true"
forest_df = pd.read_csv(url)
print(forest_df.head(10))
```

```python
forest_df.head()[["Entity", "Year", "Code","Annual net change in forest area"]]
```

```python
forest_df.describe(include='object')
```

```python
forest_df.describe(include='all')
```

```python
#AI used to help modify and narrow down enities shown 

top_entities = forest_df.groupby('Entity')['Annual net change in forest area'].mean().nlargest(10).index
top_df = forest_df[forest_df['Entity'].isin(top_entities)]


plt.figure(figsize=(14, 7))
sns.lineplot(data=top_df, x='Year', y='Annual net change in forest area', hue='Entity', marker='o')
plt.title("Annual Forest Change Rate Over Time by Entity")
plt.xlabel("Year")
plt.ylabel("Annual Forest Change Rate")
plt.legend(title='Entity', loc='upper left')
plt.show()
```

Data used to show Annual net change in forest area sorted by enitity.
